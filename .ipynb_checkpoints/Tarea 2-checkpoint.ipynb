Siga trabajando con la misma serie de tiempo, los datos trimestrales del crédito financiero en el Perú. Con ello realice pronóstico de dos trimestres adicionales, el cual debe considerar la tendencia y estacionariedad de la serie.

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima_model import ARIMA

# Cargamos los datos en un DataFrame
data = pd.DataFrame({'Trimestre': trimestres, 'Valor':valores_diferenciados})

# Identificar el modelo ARIMA usando ACF y PACF
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Gráfico de ACF y PACF
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))
plot_acf(data['Valor'], lags=15, ax=ax1)
plot_pacf(data['Valor'], lags=15, ax=ax2)
plt.show()
```

```python
# Basado en los gráficos ACF y PACF, podemos elegir un modelo ARIMA(1,0,1)
p = 1  # Orden del componente AR
d = 0  # Orden de diferenciación
q = 1  # Orden del componente MA

# Ajustar el modelo ARIMA
model = ARIMA(data['Valor'], order=(p, d, q))
model_fit = model.fit(disp=0)

# Resumen del modelo
print(model_fit.summary())

# Realizar pronóstico de 2 trimestres adicionales
forecast_steps = 2
forecast, stderr, conf_int = model_fit.forecast(steps=forecast_steps)

# Obtener el trimestre siguiente al último observado
ultimo_trimestre = data['Trimestre'].iloc[-1]
ultimo_trimestre_numero = int(ultimo_trimestre[1:])
proximos_trimestres = [f'T{ultimo_trimestre_numero + i + 1}' for i in range(forecast_steps)]

# Crear un DataFrame para el pronóstico
pronostico_df = pd.DataFrame({'Trimestre': proximos_trimestres, 'Pronóstico': forecast})

# Calcular los valores originales a partir de la diferencia
ultimo_valor_original = valores[-1]  # Último valor original en la serie
pronostico_original = np.cumsum(np.insert(forecast, 0, ultimo_valor_original))

# Agregar el pronóstico original al DataFrame
pronostico_df['Pronóstico Original'] = pronostico_original

# Mostrar el pronóstico
print(pronostico_df)
```

## Pregunta 2

Escriba la función de verosimilitud del proceso $y_{t}=c + \phi y_{t-1}+ \varepsilon_{t}$ donde $\varepsilon_{t}$ es un proceso gaussiano con media 0 y varianza $ \sigma^{^2}$. Indique también el problema de maximización a resolver.

### Solucion

El proceso presentado, por sus características, pertenece a un modelo AR(1), donde la función de verosimilitud es usada para estimar los valores más probables de los parámetros que se ajusten mejor a los datos; dichos parámetros vendrían a ser: la constante $c$, el coeficiente de autoregresión 1 y la varianza $ \sigma^{^2}$


Sabiendo que el proceso $y_{t}=c + \phi y_{t-1}+ \varepsilon_{t}$ donde $t$ es un proceso gaussiano con media 0 y varianza $ \sigma^{^2}$, se puede escribir una función de densidad de probabilidad condicional para $t\sim iid N(0,\sigma^{^2})$ como:



$$f(\varepsilon_{t}|\varepsilon_{t-1})=(2\pi\sigma^{^2})^{\frac{-1e(-\varepsilon_{t}^2)}{2(2 \sigma^{^2})}}$$

Se asumen los errores como independientes e idénticamente distribuidos (i.i.d.), además se puede agrupar a los parámetros del modelo en una constante , también se asume que el vector y de observaciones gaussiano dado que lo mismo se asumió con los errores; además, se puede usar T para simbolizar el número de observaciones. Con lo anterior descrito, se puede plantar la función de verosimilitud como: 


$$\beta =\begin{pmatrix}
c\\ 
\phi_{1}\\ 
\sigma^{^2}
\end{pmatrix}

y=\begin{pmatrix}
y_{1}\\ 
..\\ 
y_{t}
\end{pmatrix}$$

$$L(\beta|y)=\prod_{t-2}^T\frac{1}{\sqrt{2\pi\sigma^{^2}}}*e^{\frac{-(y_{t}-c\phi_{1}y_{t-1})^2}{2\sigma^{^2}}} $$

Para este caso, el problema de maximización sería encontrar los valores de los parámetros    que maximizan la función de verosimilitud. Es decir, se debe hallar los valores de $\beta=(c,\phi_{1},\sigma^{2}) $ que hacen que los datos observados sean más probables bajo el modelo dado. Este proceso se puede hacer usando el método de mínimos cuadrados o con el método de máxima verosimilitud que implica derivar la función de verosimilitud con respecto a los parámetros y encontrar los valores que maximizan esta derivada. El estimador de máxima verosimilitud para $\sigma^{2}$ será igual a la varianza muestral de los residuos, mientras que los estimadores de $𝑐$ y $\phi_{1}$ se pueden encontrar utilizando técnicas estándar de regresión lineal.


$$L(\beta|y)=f_{y}(\beta:y)\\
\widehat{\beta_{MV}}=Arg Max L(\beta|y)\\
\widehat{\beta_{MV}}=MaxLogL(\beta|y)\\
\frac{\partial }{\partial \widehat{\beta }}LogL(\beta|y)=0\rightarrow\widehat{\beta_{MV}}$$

## Pregunta 3  

Convierta el modelo $y_{t}=\phi_{1}y_{t-1}+\phi_{2}y_{t-2}+\phi_{3}y_{t-3}+\varepsilon_{t}$ en un $MA(\infty )$ usando la propiedad de invertibilidad. Mencione qué condiciones deben cumplirse para poder hacer esto y exprese también el valor de los 5 primeros coeficientes del proceso resultante.



### Solucion

La principal condición a cumplirse para que un modelo $AR(p)$ sea invertible, de tal forma que se vuelva un $MA(\infty )$, es que las raíces del polinomio de rezagos sean mayores que 1.



Esto a su vez garantiza lo siguiente:

Si $|Li|>|$ entonces $\exists[\Theta]^{-1}$ , lo que basicamente implica que existe la inversa del modelo $AR$

Ahora bien,procederemos a definir el polinomio caracteristico del modelo $AR$

$$y_{t}-\phi_{1}Ly_{t}-\phi_{2}L^2y_{t}-\phi_{3}L^3y_{t}=\varepsilon_{t}\\
(1-\phi_{1}L-\phi_{2}L^2-\phi_{3}L^3)y_{t}=\varepsilon_{t}\\
y_{t}=(\frac{1}{1-\phi_{1}L-\phi_{2}L^2-\phi_{3}L^3})\varepsilon_{t}$$

Continuamos calculando los 5 coeficientes:


$$\Psi_{0}=1 \\
\Psi_{1}= \phi_{1}\\
\Psi_{2}= \phi_{2}+\phi_{1}^2\\
\Psi_{3}= \phi_{3}+2\phi_{1}\phi_{2}+\phi_{1}^{3}\\
\Psi_{4}= 2\phi_{1}\phi_{3}+3\phi_{1}^2\phi_{2}+\phi_{1}^{4}+\phi_{2}^2\\$$
